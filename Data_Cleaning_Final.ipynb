#Libraries  and Data importing


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/user/Desktop/Dataets/Call_list.csv")
df
```

# modify Phone_Number column


```python
df['Phone_Number']=df['Phone_Number'].str.replace('|','')
df['Phone_Number']=df['Phone_Number'].str.replace('/','')
df['Phone_Number']=df['Phone_Number'].str.replace('--','')
df['Phone_Number']=df['Phone_Number'].str.replace('-','')
df['Phone_Number']
```

# Giving Phone_Number column a standard form


```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x)[0:3]+'-'+str(x)[3:6]+'-'+str(x)[6:10])
df['Phone_Number']
```

# splitting Address column and making 3 columns

```python
df[['Street','City','Zip-Code']]=df['Address'].str.split(',',2,expand=True)
df
```

# Removing uncessary columns

```python
df=df.drop(columns=['Unnamed: 8','Unnamed: 9','Address'])
df

```

# Dealing with NULL  Value

```python
df=df.fillna('')
df
```

# Change 'Yes' and 'No' to 'Y' and 'N' respectively in  'Paying Customer' and 'Do_Not_Contact columns.

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('No','N')
df['Paying Customer']=df['Do_Not_Contact'].replace('Yes','Y')
df['Paying Customer']=df['Do_Not_Contact'].replace('No','N')
df
```

```python
df=df[df['Do_Not_Contact']!='Y']
df=df[df['Phone_Number']!='nan--']
df=df[df['Phone_Number']!='Na--']
df=df.reset_index(drop=True)
df
```
